```python
import re
from deepdiff import DeepDiff
import json

json_dict = {}
lines = []

checked_option = {'0': 'Unchecked', '1': 'Checked'}
binary_option = {'1': 'Yes', '0': 'No'}
telugu_binary_option = {
      "0": "లేదు",
      "1": "అవును"
    }
binary_with_trip = {
        "0": " no ",
        "1": " yes "
    }

stupid_binary = {
    "1": "Yes",
    "2": "No"
}

with open("DISA_alldata_cleaning_english variables.txt", "r") as file_ptr:
    lines = file_ptr.readlines()


# Go through each of the lines and parse out the things we need

options_line_identification_expression = r'label\sdefine\s(\w+)_.*'
# options_expression = r'label\sdefine\s(\w+)\s+((\d+)\s+"?(\w+)"?\s+)+'
options_extraction_pattern = r'(\d+)\s+"([^"]+)"'

for line in lines:
    matches = re.match(options_line_identification_expression, line)
    if matches:
        variable_name = matches.group(1)
        var_matches = re.findall(options_extraction_pattern, line)
        print(f"Variable Name: {variable_name}, Matches: {len(var_matches)}")
        options = {}
        for old_value, new_value in var_matches:
            stripped_value = new_value.strip()
            print(old_value, stripped_value)
            options[old_value] = stripped_value
        
        # Check if this is one of the ignores
        if not DeepDiff(options, checked_option, ignore_string_case=True):
            print(f"Skipping: {variable_name}")
            print(options)
            continue
        if not DeepDiff(options, binary_option, ignore_string_case=True):
            print(f"Skipping: {variable_name}")
            print(options)
            continue
        if not DeepDiff(options, telugu_binary_option, ignore_string_case=True):
            print(f"Skipping: {variable_name}")
            print(options)
            continue
        if DeepDiff(options, stupid_binary):
            options = {
                "1": 1,
                "2": 0
            }
        json_dict[variable_name] = options


```

```python
with open("new_dict.json", "w") as file_ptr:
    json.dump(json_dict, file_ptr)
```
